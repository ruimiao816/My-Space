```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_net = pd.read_csv("/Users/hyq/Desktop/Exeter/Business_Project/DATASET_For_BEMM464J_Project.csv", encoding='utf=8', index_col = 0)
```

```python
df_net
```

```python
G = nx.DiGraph()
for i in df_net.columns:
       for j in df_net.index:
            if df_net[i][j] != 0:
                G.add_edge(j,i, weight = df_net[i][j])
plt.figure(figsize=[30,30])
nx.draw(G)                
```

```python
# Calculation of degree centrality
Gdegree=nx.degree(G)
Gdegree=dict(Gdegree)
Gdegree=pd.DataFrame({'name':list(Gdegree.keys()),'degree':list(Gdegree.values())})
Gdegree.sort_values(by="degree",ascending=False).head(15).plot(x='name',y='degree',kind='bar',figsize=(12,6),legend=False)
plt.xlabel("name",fontsize=10)
plt.ylabel("degree",fontsize=10)
plt.show()
```

```python
#Calculation of betweenness centrality
Bdegree=nx.betweenness_centrality(G)  
Bdegree=pd.DataFrame({'name':list(Bdegree.keys()),'Bdegree':list(Bdegree.values())})
Bdegree.sort_values(by="Bdegree",ascending=False).head(15).plot(x='name',y='Bdegree',kind='bar',figsize=(12,6),legend=False)
plt.xlabel("name",fontsize=10)
plt.ylabel("Bdegree",fontsize=10)
plt.show()
```
