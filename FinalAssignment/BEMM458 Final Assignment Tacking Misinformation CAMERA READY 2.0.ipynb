## Question 1. Which twitter conversations have large levels of amplification ? (15 points)
### Tip: consider number of retweets, likes, etc as proxies to measure the spread of information on social media

```python
#Find the twitter conversations which have the large number of retweets.
topretweet=spainEnglishAll.sort_values(by='public_metrics.retweet_count',ascending=False)
topretweet.head(20)
```

```python
#The result shows that the top 20 are all retweets of the conversations created by NCTsmtown_127. 
#So, I'd like to find out conversation  which is not created by NCTsmtown_127 has the larger number of retweet.
#Then I find out the following conversations.

othertopretweet=topretweet[~topretweet['tweet.text'].str.contains('NCTsmtown_127')]
othertopretweet.head(10)
```

```python
# As We can see the top-ranked are all retweets. What if we devide them and just look at how the original tweets are ranked?
#Well, we can see the first two tweets are conversations about soccer, following them are about country view, infrastructure，
#livelihood issues,cryptocurrency etc.

originaltweet=spainEnglishAll[~spainEnglishAll['tweet.text'].str.contains('RT @')]
originaltweet.sort_values(by='public_metrics.retweet_count',ascending=False).head(10)
```

```python
#Next, I'd like to refer to the number of replys and likes, so that I can make a comprehensive evaluation from multiple factors.
#First, find out which conversations have the larger number of replys.
spainEnglishAll.sort_values(by='public_metrics.reply_count',ascending=False).head(10)
```

```python
#Second, find out which conversations have the larger number of likes.
spainEnglishAll.sort_values(by='public_metrics.like_count',ascending=False).head(10)
```

```python
#As shown in the dataframe, conversation "EURO 2008 top scorer: Villa..." ranks the first place in all the three factors:retweet, replys and likes.So, it can be said that this conversation is the most amplified one of the original tweets.
#And the conversation ""What if Lionel Messi... " ranks the second in retweet and likes, the fourth in replys.Overall, it can be considered as the second most amplified conversation of the original tweets.
#The third most amplified conversation of the original tweets can be given to the "From Spain: Manchester...".Since it ranks the second in replys, the third in likes and the seventh in retweets.
```
