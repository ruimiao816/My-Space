```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
y = [7119,7486,7969,8278,8565,8756,9152,9408,9335,8981,7803]
z = [0.02,0.05,0.06,0.04,0.03,0.02,0.05,0.03,-0.01,-0.04,-0.13]
fig, ax = plt.subplots(1, figsize=(16,9))#figsize是比例，可自行调整

ax1 = sns.barplot(x,y,
    palette='deep',  #palette是颜色深浅
    ax=ax)

ax.set_ylim([0, 10000]) #设定y轴数值大小
ax.set_ylabel("GLOBAL SALES", fontsize=18) #fontsize是字体大小
ax.set_xlabel("YEAR", fontsize=18)
ax.set_title("Annual Sales from 2010 to 2020", fontsize=18)

sns.despine(offset=15)#despine是设定图的上下眉脚的宽度
xticks = ax.get_xticks()
for i in range(len(y)):
    xy = (xticks[i], y[i] * 1.03)
    s = str(y[i])
    ax.annotate(
        s=s,  # 要添加的文本
        xy=xy,  # 将文本添加到哪个位置
        fontsize=12,  # 标签大小
        color="blue",  # 标签颜色
        ha="center",  # 水平对齐
        va="baseline"  # 垂直对齐
    )


plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(16,9))#figsize是比例，可自行调整

plt.plot(x,z,'r',ms=10,lw=2,marker='o')
plt.ylabel("GROWTH RATE", fontsize=18)


sns.despine(offset=15)#despine是设定图的上下眉脚的宽度




plt.show()
```

```python
PP=pd.read_csv('PP.csv')
PP.head()
```

```python
PP["proportion"] = pd.to_numeric(PP["proportion"],errors='coerce')
```

```python
PP.info()
```

```python
fig, ax = plt.subplots(1, figsize=(14,7))#figsize是比例，可自行调整

ax = sns.boxplot(
    x='region',
    y='proportion',
    data=PP,
    fliersize=0,
    palette='deep',  #palette是颜色深浅
    ax=ax)

ax.set_ylim([0.04, 0.28]) #设定y轴数值大小
ax.set_ylabel("PROPORTION", fontsize=18) #fontsize是字体大小
ax.set_xlabel("REGIONS",fontsize=18)
ax.set_title("Proportion of Auto Sales in Different Regions",fontsize=20)

sns.despine(offset=20)#despine是设定图的上下眉脚的宽度


plt.show()
```

```python
df=pd.read_csv('CarPrice_Assignment-吉利.csv')
```

```python
df
```

```python
df.info()
```

```python
df.drop(['car_ID'],axis = 1,inplace = True)
df
```

```python
cat = ['fueltype', 'aspiration', 'doornumber','drivewheel', 'enginelocation','enginetype','cylindernumber','fuelsystem']
print('No. of categorical columns are:',len(cat))
```

```python
num = []
for i in df.columns:
    if i not in cat:
        num.append(i)
print('No. of numerical/float columns are:',len(num))
```

```python
df['fueltype'].value_counts()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15,8))
x=df['fueltype'].value_counts().index
y=df['fueltype'].value_counts().values.tolist()
data = df.groupby("fueltype").size()
sns.set(style="dark", color_codes=True)
pal = sns.color_palette("magma", len(data))
rank = data.argsort().argsort() 
sns.barplot(x=x,y=y,palette=np.array(pal[::-1])[rank],ax = ax[0])
for p in ax[0].patches:
        ax[0].annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.3, p.get_height()),
                    ha='center', va='bottom',
                    color= 'black')
ax[0].set_xlabel('fueltype (gas/diesel)', weight='semibold', fontname = 'monospace')
_, _, autotexts= ax[1].pie(y, labels = x, colors = pal, autopct='%1.1f%%',
        explode=[0.03 for i in df['fueltype'].value_counts().index])
for autotext in autotexts:
    autotext.set_color('white')
plt.legend(bbox_to_anchor=(1, 1))
plt.suptitle ('Fueltype(gas/diesel)',weight = 'bold')
plt.show()
```
