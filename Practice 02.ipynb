```python
AQCities=pd.concat([Beijing,Berlin,London,MexicoDC,NewDelhi],axis=0)
```

```python
AQCities
```

```python
AQCitiesco=AQCities[AQCities['parameter']=='co']
```

```python
AQCitiesco.sample(10)
```

```python
fig,ax=plt.subplots(1,figsize=(16,9))
ax=sns.boxplot(x='city',y='value',data=AQCitiesco,fliersize=0,palette='deep',ax=ax)
ax.set_title('Co value of Cities');
ax.set_xlabel('city');
ax.set_ylim([0,750])
ax.set_ylabel('co',fontsize=18);
sns.despine(offset=10)
plt.show()
```

```python
AQcitiesPM25=AQCities[AQCities['parameter']=='pm25']
AQcitiesPM25
```

```python
fig,ax=plt.subplots(1,figsize=(14,7))
ax=sns.boxplot(x='city',y='value',data=AQcitiesPM25,fliersize=0,palette='deep',ax=ax)
ax.set_title('PM25 value of Cities');
ax.set_xlabel('city');
ax.set_ylim([0,750])
ax.set_ylabel('pm25',fontsize=18);
sns.despine(offset=10)
plt.show()
```

```python
import pandas as pd
```

```python
air_quality_no2 = pd.read_csv('https://raw.githubusercontent.com/thousandoaks/BEMM458_OLD/master/data/air_quality_no2_long.csv',parse_dates=True)
```

```python
air_quality_pm25 = pd.read_csv('https://raw.githubusercontent.com/thousandoaks/BEMM458_OLD/master/data/air_quality_pm25_long.csv',parse_dates=True)
```

```python
air_quality_no2.head()
```

```python
air_quality_pm25.head()
```

```python
airquality=pd.concat([air_quality_no2,air_quality_pm25],axis=0)
```

```python
airquality
```

```python
air_quality_parameters = pd.read_csv("https://raw.githubusercontent.com/thousandoaks/BEMM458_OLD/master/data/air_quality_parameters.csv")
```

```python
air_quality_parameters
```

```python
stations_coord = pd.read_csv("https://raw.githubusercontent.com/thousandoaks/BEMM458_OLD/master/data/air_quality_stations.csv")
```

```python
stations_coord.head()
```

```python
airquality_geolocation=pd.merge(airquality,stations_coord,how='left',on='location')
```

```python
airquality_geolocation
```

```python
airquality_complete=pd.merge(airquality_geolocation,air_quality_parameters,how='left',left_on='parameter',right_on='id')
```
