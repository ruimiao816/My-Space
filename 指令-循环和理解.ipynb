#while循环

-无终止符，只要条件为真就一直循环

```python
i = 1
while i < 6:
  print(i)
 
  i += 1
```

*条件i<6

-有终止指令，退出循环

```python
i = 1
while i < 300:
  print(i)
  if i == 12:
    print('found 12, I exit')
    break
  i += 1
```

#for循环

for循环用来迭代序列（即列表、元组、字典、集合或字符串）

```python
fruits = ["apple", "banana", "cherry","tomato","pomegranate","orange"]
for element in fruits:
  print(element)
```

```python
for element in fruits:
    elementupper=element.upper()
    print(elementupper)
```

-循环终止指令

```python
for element in fruits:
    print(element)
    if element == 'tomato':
        print('Found the tomato, I exit')
        break
```

-选择产生新序列

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango","tomato","orange","berry"]
newlist = []

for element in fruits:
  if "a" in element:
    newlist.append(element)

print(newlist)
```

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango","tomato","orange","berry"]
newlist = []

for item in fruits:
  if "y" in item:
    newlist.append(item)

print(newlist)
```

-使用 range() 函数指定代码循环次数

```python
for x in range(0,20):
  print("I am at:",x)
```
