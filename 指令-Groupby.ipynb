#Groupby指令

-载入数据

```python
图表名=pd.read_csv('链接',seq='\t')
```

-挑选列数据

```python
图表名['表头名']
图表名[['表头名1','表头名'2]]
```

-挑选行数据

```python
rowstoselect=[行数1，行数2，行数3]
rowstoselect=[行数1:行数2]
图表名.iloc[rowstoselect]
```

-Groupby

```python
图表名.groupby('筛选背景列')['筛选结果列']
图表名.groupby(['筛选背景列1','筛选背景列2'])['筛选结果列']
后续跟计算:平均值.mean；计数.nunique；合计.sum；
```

-Groupby列举

```python
pd.set_option('display.max_rows',最大展示行数)
图表名.groupby('筛选背景列1','筛选背景列2')['筛选结果列'].计算().head(最大展示行数)
```

-Groupby后做图

```python
单筛选背景+结果列
图表名.plot()
```

```python
多筛选背景+结果列
图表名.unstack(level=0).plot(kind='line', subplots=False,figsize=(自定义,自定义))
```

-Groupby后排序

正向值排序

```python
图表名.sort_values(by='序列表头名',ascending=True)
```

反向值排序

```python
图表名.sort_values(by='序列表头名',ascending=False)
```

```python
图表名.sort_index(axis=1, ascending=False)
```
