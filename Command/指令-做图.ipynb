#做图

-sns载入数据

```python
图表名=sns.load_dataset("数据名")
```

-Histogram直方图

```python
ax = sns.distplot(图表名['列表头'])
ax.set_title('自定义名称');
```

-Bar plot柱状图

```python
ax = sns.countplot('列表头', data=图表名)
ax.set_title('自定义名称');
ax.set_xlabel('Day of the Week');
ax.set_ylabel('Frequency');
```

-Scatter plot散点图

```python
ax = sns.regplot(x='关联列1', y='关联列2', data=图表名)
ax.set_title('自定义名称');
ax.set_xlabel('关联列1名称');
ax.set_ylabel('关联列2名称');
```

-Density plot密度图

```python
ax = sns.kdeplot(data=图表名['列表头1'],data2=tips['列表图2'],shade=True) 
ax.set_title('自定义名称');
ax.set_xlabel('列表头1');
```

-Box plot箱体图

```python
ax = sns.boxplot(x='列表头1', y='列表头2', data=图表名)
ax.set_title('自定义')
ax.set_xlabel('列表头1')
ax.set_ylabel('列表头2')
```

-Violin plot提琴图

单色提琴图

```python
ax = sns.violinplot(x='列表头1', y='列表头2', data=图表名)
ax.set_title('自定义')
ax.set_xlabel('列表头1')
ax.set_ylabel('列表头2')
```

颜色区分提琴图

```python
ax = sns.violinplot(x='列表头1', y='列表头2',hue='颜色区分条件', data=图表名,split=True)
```

-Pair plot组合图

简单组合图

```python
fig = sns.pairplot(图表名)
```

复杂组合图

```python
pair_grid = sns.PairGrid(图表名)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug=False)
plt.show()
```

颜色区分组合图

```python
fig = sns.pairplot(tips, hue='颜色区分条件')
```

-Facet plot

```python
facet = sns.FacetGrid(tips, col='条件表头', hue='颜色区分条件')
facet = facet.map(plt.scatter, '结果列表头1', '结果列表头2')
facet = facet.add_legend()
```

-多条件设置做图

```python
fig, ax = plt.subplots(1, figsize=(14,7))#figsize是比例，可自行调整

ax = sns.boxplot(
    x='筛选条件',
    y='数值项',
    data=图表名,
    fliersize=0,
    palette='deep',  #palette是颜色深浅
    ax=ax)

ax.set_ylim([0, 750]) #设定y轴数值大小
ax.set_ylabel("设定y轴显示名称", fontsize=18) #fontsize是字体大小
ax.set_xlabel("设定x轴显示名称")

sns.despine(offset=10)#despine是设定图的上下眉脚的宽度


plt.show()
```
