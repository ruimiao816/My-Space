-Filter

String精确筛选

```python
xxFilter=图表名['筛选列表头']=='筛选条件值'
```

```python
图表名[xxFilter]
```

String模糊筛选

```python
xxFilter=图表名['筛选列表头'].str.contains('筛选条件值')
```

```python
图表名[xxFilter]
```

单项数值条件筛选

```python
xxFilter=图表名['筛选列表头']>=<数值
```

```python
图表名[xxFilter]
```

多项数值条件筛选

```python
xxFilter=（图表名['筛选列表头']>=<数值1）&（图表名['筛选列表头']>=<数值2）
```

```python
图表名[xxFilter]
```

一次筛选某列多个项

```python
新图表名=图表名[(图表名['表头名']=='筛选值1')|(图表名['表头名']=='筛选值2')|(图表名['表头名']=='筛选值3')]
```

```python
新图表名
```
